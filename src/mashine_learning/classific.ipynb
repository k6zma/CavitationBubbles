```python
import numpy as np
import pandas as pd
from sklearn.utils import shuffle
```

```python
train_df = pd.read_csv('../../data/fixed_data/ResultTrain_FixedData.csv')
train_df
```

```python
train_df = shuffle(train_df)
train_df
```

```python
test_df = pd.read_csv('../../data/fixed_data/ResultTest_FixedData.csv')
test_df
```

```python
test_df = shuffle(train_df)
test_df
```

```python
X_train = train_df.drop(['Concentration','Photo Name'], axis=1)
X_train
```

```python
y_train = train_df[train_df.columns[1]].astype('int')
y_train
```

```python
X_test = test_df.drop(['Concentration', 'Photo Name'], axis=1)
X_test
```

```python
y_test = test_df[test_df.columns[1]].astype('int')
y_test
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
nb_classifier = GaussianNB()

params_NB = {'var_smoothing': np.logspace(0,-9, num=100)}
GaussianNB_model = GridSearchCV(estimator=nb_classifier, 
                 param_grid=params_NB, 
                 cv=7,   # use any cross validation technique 
                 verbose=1, 
                 scoring='accuracy') 
GaussianNB_model.fit(X_train,y_train) 
```

```python
pred_train_GaussianNB = GaussianNB_model.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test.values, pred_train_GaussianNB))
```
